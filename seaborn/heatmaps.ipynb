```python
import numpy as np
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.describe()
```

```python
flights.info()
```

```python
tc = tips.corr()
```

```python
tc
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp
```

```python
# cmap argument references the colors of the heatmap
# linecolor argument references the color of the line to separate the elements
# linewidths argument references how thick of a line to separate the elements

sns.heatmap(fp, cmap='coolwarm', linecolor='black', linewidths=1)
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
