````python
import seaborn as sns
sns.set_style("whitegrid")
````

````python
%matplotlib inline
````

````python
iris = sns.load_dataset("iris")
````

````python
iris.info()
````

````python
tips = sns.load_dataset("tips")
````

````python
tips.info()
````

### single horizontal boxplot

````python
sns.boxplot(x=tips["total_bill"])
````

### vertical boxplot grouped by a categorical variable

````python
sns.boxplot(x="day", y="total_bill", data=tips)
````

### boxplot with nested grouping by two categorical variables

````python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
````

### boxplot with nested grouping when some bins are empty

````python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips, linewidth=2.5)
````

#### control box order by passing an explicit order

````python
sns.boxplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
````

### boxplot for each numeric variable in a DataFrame

````python
sns.boxplot(data=iris, orient="h", palette="Set2")
````

### use ```hue``` without changing box position or width

````python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
````

````python
sns.boxplot(x="day", y="total_bill", hue="weekend", data=tips)
````

### use swarmplot() to show the datapoints on top of the boxes

````python
sns.boxplot(x="day", y="total_bill", data=tips)
````

````python
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
````

````python
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
````

## use factorplot() to combine a boxplot() and a FacetGrid

````python
# This allows grouping within additional categorical variables. 
# Using factorplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets
````

````python
sns.factorplot(x="sex", y="total_bill",
               hue="smoker", col="time",
               data=tips, kind="box",
               size=4, aspect=.7);
````

### grouped boxplots
