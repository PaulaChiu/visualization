```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
tips.describe()
```

```python
tips.info()
```

## histogram

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=40)
```

## jointplot

```python
# kind=hexcreates a hexagon plot that has darker colors for high density areas

sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# defaultcreates a scatter plot with individual points

sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# kind=regcreates a regression line to display the linear fit

sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# kind=kdedisplays the density of where the points line up the most

sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

## pairplot

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```
